```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By

driver = webdriver.Edge()
driver.get("https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue")
driver.maximize_window()
time.sleep(2)
```

```python
# Locating the second table using XPath

target_table = driver.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/table[2]")
print("Second table located.")
```

```python
rows = target_table.find_elements(By.TAG_NAME, "tr")[1:]  # skip header
```

```python
# Looping through Rows to print all columns


for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = [cell.text.strip() for cell in cells]
    print(row_data)
```

```python
import pandas as pd

# Collect all rows into a list
data = []
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = [cell.text.strip() for cell in cells]
    if row_data:  # skip empty rows
        data.append(row_data)

# Create DataFrame
columns = ["Company", "Revenue (USD billions)", "Profit (USD billions)", "Employees", "Headquarters"]
df = pd.DataFrame(data, columns=columns)

# Preview
df.head()
```

```python
# Saving df as CSV
df.to_csv("tech_companies_by_revenue.csv", index=False)
print("CSV file saved successfully.")
```
