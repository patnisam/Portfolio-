# Amazon Cell Phones Reviews Sentiment Analysis.

### - Sentiment Scoring of Amazon Cellphones Reviews.
##### Dataset: https://www.kaggle.com/datasets/grikomsn/amazon-cell-phones-reviews?select=20191226-reviews.csv

#### Importing Relevant Libraries

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

import nltk
nltk.download('stopwords')
import nltk
nltk.download('punkt_tab')
import nltk
nltk.download('wordnet')
```

#### Loading Dataset
 

##### -Using 'body' column/feature which has reviews of the cell phones.

```python
df = pd.read_csv('20191226-reviews.csv',usecols = ['body'])
```

#### Building Sentiment Scoring Model

##### Introducing Word Net Lemmatizer and English Stopwards

```python
lemma = WordNetLemmatizer()
stop_words = stopwords.words('english')
```

#### Function to preprocess the text

###### 1. Convert the input string to lowercase.

###### 2. Extraction of letters from only this string using Regular Expressions.

###### 3. Tokenize the string.

###### 4. Filter the text of stop words.

###### 5. Lemmatize the filtered words and return them.

```python
def prep_text(x):
    corpus = str(x).lower()
    corpus = re.sub('[^a-zA-Z]+',' ', corpus).strip()
    tokens = word_tokenize(corpus)
    words = [t for t in tokens if t not in stop_words]
    lemmatize = [lemma.lemmatize(w) for w in words]

    return lemmatize
```

##### Application of the Function.

```python
preprocess_tag = [prep_text(i) for i in df['body']]
df['preprocess_txt'] = preprocess_tag
```

##### Calculate the number of filtered words for each text

```python
df['total_len'] = df['preprocess_txt'].map(lambda x: len(x))
```

##### Defining the generic Negative Words and Positive Words list. 

##### Using the positive-text.txt and negative-text.txt files, also known as Opinion Lexicon.
##### Link to files: https://www.cs.uic.edu/~liub/FBS/sentiment-analysis.html

```python
file = open('negative-words.txt', 'r')
neg_words = file.read().split()
file = open('positive-words.txt', 'r')
pos_words = file.read().split()
```

#### Counting positive and negative words for the text.

##### Words that are present in the positive words (pos_words) and all the words in preprocess_txt that are present in the negative words (neg_word).

```python
num_pos = df['preprocess_txt'].map(lambda x: len([i for i in x if i in pos_words]))
df['pos_count'] = num_pos
num_neg = df['preprocess_txt'].map(lambda x: len([i for i in x if i in neg_words]))
df['neg_count'] = num_neg
```

```python
df['sentiment'] = round((df['pos_count'] - df['neg_count']) / df['total_len'], 2)
```
