```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Setting headers to avoid bot block

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

url = "https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
tables = soup.find_all("table", class_="wikitable")
target_table = tables[1]  # second table
```

```python
# Extracting headers
headers = [th.text.strip() for th in target_table.find_all("tr")[0].find_all("th")]

# Extracting rows
data = []
for row in target_table.find_all("tr")[1:]:
    cells = row.find_all("td")
    row_data = [cell.text.strip() for cell in cells]
    if row_data:
        data.append(row_data)
```

```python
df = pd.DataFrame(data, columns=headers)
df.to_csv("tech_companies_bs.csv", index=False)
print("CSV saved successfully.")
```
